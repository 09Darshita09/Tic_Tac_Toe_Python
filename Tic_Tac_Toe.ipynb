# Tic Tac Toe Game

The steps to play the game are:

1. Player 1 gets to choose his/her symbol.
2. Player will be asked to be ready to play.
3. Randomly either Player 1 or 2 goes first.
4. The player who has the turn should choose a number between 1 to 9 to place his/her respective symbol on the board 



5. Then the next player must do the same.
6. Player can either win, lose or tie with each other
7. After the results the players are asked if they want to replay or not.


```python
#For input from the player for the symbol
def choose_symbol():
    choice = ' '
    symbol = ['X','O']
    
    while choice.upper() not in symbol :
        choice = (input('Player 1, please choose your symbol (X or O): ')).upper()
    if choice == 'X':
        return ('X','O')
    else:
        return ('O','X')
```

```python
#For input from the player if ready or not
def ready():
    choice = ' '
    opinion = ['Y','N']
    
    while choice.upper() not in opinion :
        choice = (input('Are you ready to play? (Y or N): ')).upper()
    return choice  
```

```python
#Function to choose randomly which player goes first
import random
def display_player():
    player = ['Player 1', 'Player 2']
    
    result = random.choice(player)
    print (result + ' will go first')
    return result
```

```python
#For user input if player wants to play again
from IPython.display import clear_output
def display_board(mark,player1_choice,player2_choice):
    clear_output()
    print ('Player 1 : ' + player1_choice +', Player 2 : ' + player2_choice)
        
    print('~~~~~~~~~~~~~')
    
    print('|',mark[6],'|',mark[7],'|',mark[8],'|')
    print('|---|---|---|')
    print('|',mark[3],'|',mark[4],'|',mark[5],'|')
    print('|---|---|---|')
    print('|',mark[0],'|',mark[1],'|',mark[2],'|')
    
    print('~~~~~~~~~~~~~') 
```

```python
#User input to choose a position on the board
def choose_position(player):
    position = 'wrong'
    within_range = False
    
    while position.isdigit() == False or within_range == False:
        position = input(player + ', Choose your next position (1,9): ')
        
        if position.isdigit() == True: 
            if int(position) in range(1,10):
                within_range = True
            else:
                within_range = False
        
    return int(position) 
```

```python
#Function to mark the user choice on board
def convert_position(board,position,choice):
    if board[position-1] == ' ':
        board[position-1] = choice
    else:
        return 'Position already filled!'

    return board
```

```python
#Function to set the winning situations
def win(mark):
   # result = False
    if mark[0] == mark[1] == mark[2] != ' '  : return True
    if mark[3] == mark[4] == mark[5] != ' '  : return True
    if mark[6] == mark[7] == mark[8] != ' '  : return True
    if mark[0] == mark[3] == mark[6] != ' '  : return True
    if mark[1] == mark[4] == mark[7] != ' '  : return True
    if mark[2] == mark[5] == mark[8] != ' '  : return True
    if mark[0] == mark[4] == mark[8] != ' '  : return True
    if mark[2] == mark[4] == mark[6] != ' '  : return True
    else: return False
```

```python
#Function for user input if player wants to play again
def replay():
    choice = ' '
    opinion = ['Y','N']
    
    while choice.upper() not in opinion :
        choice = (input('Do you want to play again? (Y or N): ')).upper() 
    return choice.startswith('Y')
```

```python
### 

print('Welcome to the TIC TAC TOE game!')

while True: 
    play = False                                      # Game is not ON
    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']     #Default board
    
    player1_choice,player2_choice = choose_symbol()                          #Choose X or O

    turn = display_player()                          #Display who goes first

    start = ready()                                   #Are you ready to play
    if start == 'Y':
        play = True
        display_board(board,player1_choice,player2_choice)
    elif start == 'N':
        play = False
        print('Ok see you next time!')
        break 
        
    while play :
        if turn == 'Player 1' :                                 
            position = choose_position(turn)                  #Ask for position
            mark = convert_position(board,position,player1_choice)   #Place position on board
            if type(mark) == str:
                print(mark)                           #Display board
            else:
                display_board(mark,player1_choice,player2_choice)
                if win(mark): 
                    print('Congratulation!! Player 1 wins')
                    play = False
                else:
                    turn = 'Player 2'
        
        elif turn == 'Player 2':                               
            position = choose_position(turn)                  #Ask for position
            mark = convert_position(board,position,player2_choice)   #Place position on board
            if type(mark) == str:
                print(mark)                           #Display board
            else:
                display_board(mark,player1_choice,player2_choice)                           #Display board
                if win(mark): 
                    print('Congratulation!! Player 2 wins')
                    play = False
                else: 
                    turn = 'Player 1'
         
        if ' ' not in mark:
            print('Game over! It is a draw game')
            play = False
    
    if not replay():
        print('Ok see you next time!')
        break
```
